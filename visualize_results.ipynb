```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def title_format(t):
    prompt_no = t.split("_")[-1][0]
    if "exp1" in t:
        if "easy" in t:
            return f"Experiment 1: Easy Prompt {prompt_no}"
        else:
            return f"Experiment 1: Medium Prompt {prompt_no}"
    else:
        return f"Experiment 2: Hard Prompt {prompt_no}"
```

```python
# Get the list of files in the directory
files = [file for file in os.listdir("data/") if "easy" in file]

# Calculate the number of rows and columns for the grid
num_plots = len(files)
num_cols = 2  # Number of columns in the grid
num_rows = -(-num_plots // num_cols)  # Ceiling division to calculate number of rows

# Create the grid of subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5*num_rows))

# Flatten the axes array if it's a single row
if num_rows == 1:
    axes = [axes]

# Iterate over files and draw plots
for i, file in enumerate(files):
    row = i // num_cols
    col = i % num_cols

    # Read the data
    d = pd.read_csv(os.path.join("data", file), index_col=0)

    # Draw the bar plot on the corresponding subplot
    sns.barplot(data=d, x="prompt", y="Surprisal", hue="Word", ax=axes[row][col], alpha=0.6, palette="dark")
    
    # Set axis labels and legend title
    axes[row][col].set_xlabel("")
    axes[row][col].set_ylabel("Surprisal")
    axes[row][col].legend(title="")
    
    # Set title for subplot
    axes[row][col].set_title(title_format(file))

# Remove empty subplots
for i in range(num_plots, num_rows * num_cols):
    row = i // num_cols
    col = i % num_cols
    fig.delaxes(axes[row][col])

plt.tight_layout()

plt.savefig("Figures/exp1_easy.png")

plt.show()
```
